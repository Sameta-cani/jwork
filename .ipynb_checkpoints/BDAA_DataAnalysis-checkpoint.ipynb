```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 로드
df = pd.read_csv('./datasets/sales_data.csv', encoding='unicode-escape')

df
```

'InvoiceDate' 칼럼이 'object'이라 'datetime'으로 형변환

```python
# 데이터의 요약통계량
df.info()
```

```python
# 'object' -> 'datetime'
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

df의 null 값 개수 확인

```python
df.isna().sum()
```

Description와 CustomerID에서의 null 값을 발견했다. 현재로서는 null값을 따로 처리할 방법이 없기에 0으로 치환한다.

```python
df['CustomerID'] = df['CustomerID'].fillna(0)
```

df의 index를 MultiIndex로 설정('StockCode', 'InvoiceDate')

```python
df.set_index(['StockCode', 'InvoiceDate'], inplace=True)
df.sort_index(level=0, inplace=True)
```

'Description' 칼럼을 보면 null값이 1,454개 존재하는데, 유심히 보니 해당 칼럼의 null값을 처리할 방법이 있다. 바로 'StockCode'가 같은 데이터의 'Description'을 그대로 붙여넣어주면 웬만한 결측치 처리는 가능할 것 같다

```python
df.isna().sum()
```

예를 들어 다음은 'Description'이 결측치인 데이터만 추출한 것인데

```python
df[df['Description'].isna()]
```

'StockCode'가 '10002'인 데이터를 추출해보자.
